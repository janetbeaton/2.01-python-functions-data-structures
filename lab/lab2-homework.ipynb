## Task 1. Validate Records

Requirements:

* Input must be a list
* Each item must be a dictionary
* Raise appropriate errors if invalid

```python
def validate_records(records):   #created a function to make sure you have a list of dictionaries
    # TODO: check type and structure
    if not isinstance(records, list):
        # we're checking to see that values is a list
        raise TypeError("values must be a list!")
    for item in records:
        if not isinstance(item, dict):
            raise TypeError("items inside the list must be a dictionary!")
    return
```

---

## Task 2. Clean Names

Requirements:

* Remove records where name is empty ("")
* Return cleaned list

```python
def clean_names(input_data):
    data = []
    for item in input_data: 
        if not len(item["name"])==0:
            data.append(item)
    return data
    
```

---

## Task 3. Handle Missing Values

Requirements:

* Replace missing score with the average score
* Replace missing hours_studied with the average hours

```python
def fill_missing_values(records):
    # TODO: calculate averages and fill missing values
    score_in=[]
    hours_in=[]
    sum_score = 0
    sum_hours=0
    length_score=0
    length_hours=0
    for item in records:
        if isinstance(item["score"],int):
            sum_score= sum_score+item["score"]
            length_score=length_score+1
        if isinstance(item["hours_studied"],int):
            sum_hours= sum_hours+item["hours_studied"]
            length_hours=length_hours+1
    average_hours=sum_hours/length_hours    
    average_score=sum_score/length_score
    
#now that i calculated the average of score and hours i can put them into the list

    for item in records:
        if isinstance(item["score"],int):
            score_in.append(item)
        else:
            item["score"]=average_score
            score_in.append(item)     
            
    for item in score_in:            
        if isinstance(item["hours_studied"],int):
            hours_in.append(item)
        else:
            item["hours_studied"]=average_hours
            hours_in.append(item)
    return hours_in
```

---

## Task 4. Fix Invalid Scores

Requirements:

* Ensure all scores are between 0 and 100
* Any score above 100 should be set to 100

```python
def cap_scores(records):
    # TODO: ensure scores are between 0 and 100
    score=[]
    for item in records:
        if isinstance(item["score"],int):
            score_in = item["score"]
            # print(score_in)
            if (score_in>100):
                item["score"]=100
        score.append(item)
    return score
```

---

## Task 5. Convert to DataFrame

Requirements:

* Convert cleaned records into a Pandas DataFrame

**MAKE SURE YOU HAVE IMPORTED PANDAS** 

```python
def to_dataframe(records):
    # TODO: convert to a DataFrame
    import pandas as pd
    df=pd.DataFrame(records)
    return df
```

---

## Task 6. Full Pipeline

Requirements:

* Call all previous functions in order
* Return final cleaned DataFrame

```python
def clean_dataset(input_data):
    # TODO: call all functions in order
    validate_records(input_data)
    clean_data=clean_names(input_data)
    filled_data=fill_missing_values(clean_data)
    capped_data=cap_scores(filled_data)   
    super_clean_data=to_dataframe(capped_data)
    
    return super_clean_data
```

```python
records = [
    {"name": "Ava", "score": 95, "hours_studied": 10},
    {"name": "Ben", "score": None, "hours_studied": 8},
    {"name": "Cara", "score": 88, "hours_studied": None},
    {"name": "", "score": 72, "hours_studied": 6},
    {"name": "Eli", "score": 105, "hours_studied": 9},
    {"name": "Finn", "score": 65, "hours_studied": 7}]

records=clean_dataset(records)
print(records)    
```
